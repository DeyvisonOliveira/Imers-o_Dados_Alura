```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Deyvison", "Carlos"],
    'salario': [4000, np.nan, 6000, np.nan]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios ['salario'].mean().round(2))

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios ['salario'].median())

df_salarios
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Deyvison", "Carlos", "Val"],
    'salario': [4000, np.nan, 6000, np.nan, 10000]
})
#Calcula a media salarial e substitui os nulos pela mdia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios ['salario'].mean().round(2))
'''
Calcula mediana e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios ['salario'].median())

df_salarios
```

# **AULA 3 - Visualizao de Dados**

```python
df_limpo.head()
```

```python
df_limpo['experience_level'].value_counts().plot(kind='bar', title="Senioridade")
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
plt.title("Salario medio por Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salrio medio anual (usd)")
plt.show()
```

```python
df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd', order=ordem)
plt.title("Salario medio por Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salrio medio anual (usd)")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.histplot(df_limpo['salary_in_usd'], bins=50, kde=True)
plt.title("Distribuio dos salarios anuais")
plt.xlabel("Salario em USD")
plt.ylabel("Frequencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['salary_in_usd'])
plt.title("Boxplot Salario")
plt.xlabel("Salario em USD")
plt.show()
```

```python
ordem_senioridade = ['EN', 'MI', 'SE', 'EX']

plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df_limpo, order=ordem_senioridade)
plt.title("Boxplot de distribuio de Salario")
plt.xlabel("Salario em USD")
plt.show()
```

```python
ordem_senioridade = ['EN', 'MI', 'SE', 'EX']

plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='experience_level')
plt.title("Boxplot de distribuio de Salario")
plt.xlabel("Salario em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['remote_ratio', 'count']

fig = px.pie(remoto_contagem,
             values='count',
             names='remote_ratio',
             title='Proporo por Senioridade')

fig.show()
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['remote_ratio', 'count']

fig = px.pie(remoto_contagem,
             values='count',
             names='remote_ratio',
             title='Proporo por Senioridade',
            hole=0.4)

fig.show()
```

```python
remoto_contagem = df_limpo["remote_ratio"].value_counts().reset_index()
remoto_contagem.columns = ["remote_ratio", "quantidade"]

fig = px.pie(remoto_contagem,
             names='remote_ratio',
             values='quantidade',
             title='proporo dos tipos de trabalho',
             hole=0.5)
fig.update_traces(textinfo="percent+label")
fig.show()
```
